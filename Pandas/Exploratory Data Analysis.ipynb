```python
# Select only numeric columns along with the 'Continent' column
numeric_columns = df.select_dtypes(include=[np.number]).columns
columns_to_keep = numeric_columns.tolist() + ['Continent']

# Create a new DataFrame with only the selected columns
df_numeric = df[columns_to_keep]

# Perform the groupby operation and calculate the mean
grouped_mean = df_numeric.groupby('Continent').mean().sort_values(by='2022 Population', ascending=False)

grouped_mean
```
