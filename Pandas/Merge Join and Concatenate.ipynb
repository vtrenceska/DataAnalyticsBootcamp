# Merge, Join and Concatenate

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"D:\Data Analytics Bootcamp\Pandas\LOTR.csv")
df1
```

```python
df2 = pd.read_csv(r"D:\Data Analytics Bootcamp\Pandas\LOTR 2.csv")
df2
```

# Merge

```python
df1.merge(df2)
```

```python
df1.merge(df2, how = 'inner', on = 'FellowshipID')
```

```python
df1.merge(df2, how = 'inner', on = ['FellowshipID','FirstName'])
```

```python
df1.merge(df2, how = 'outer')
```

```python
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'right')
```

```python
df1.merge(df2, how = 'cross')
```

# Joins

```python
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_Left', rsuffix = 'Right')
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left', rsuffix = '_Right')
df4
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left', rsuffix = '_Right', how = 'outer')
df4
```

# Concatenate

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], join = 'inner')
```

```python
pd.concat([df1,df2], join = 'outer')
```

```python
pd.concat([df1,df2], join = 'inner', axis = 1)
```

```python
# df1.append(df2) - does not work, instead use concat
```

```python
df1 = pd.DataFrame(df1)
df2 = pd.DataFrame(df2)

combined_df = pd.concat([df1, df2])

combined_df
```
