# Group By and Aggregating

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("D:\Data Analytics Bootcamp\Pandas\Flavors.csv")
```

```python
df
```

```python
df.groupby('Base Flavor')
```

```python
group_by_frame = df.groupby('Base Flavor')
```

```python
#group_by_frame.mean()
```

```python
# Group by 'Base Flavor'
group_by_frame = df.groupby('Base Flavor')

# Select numeric columns
numeric_cols = df.select_dtypes(include=[np.number]).columns

# Calculate mean for numeric columns
mean_values = group_by_frame[numeric_cols].mean()

print(mean_values)
```

```python
df.groupby('Base Flavor').count()
```

```python
df.groupby('Base Flavor').min()
```

```python
df.groupby('Base Flavor').max()
```

```python
df.groupby('Base Flavor').sum()
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating': ['mean','max','count','sum']})
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating': ['mean','max','count','sum'], 'Texture Rating': ['mean','max','count','sum']})
```

```python
df.groupby(['Base Flavor','Liked'])
```

```python
#df.groupby(['Base Flavor','Liked']).mean()
```

```python
# Group by 'Base Flavor' and 'Liked'
group_by_frame = df.groupby(['Base Flavor', 'Liked'])

# Select numeric columns
numeric_cols = df.select_dtypes(include=['number']).columns

# Calculate mean for numeric columns
mean_values = group_by_frame[numeric_cols].mean()

mean_values
```

```python
df.groupby(['Base Flavor','Liked']).agg({'Flavor Rating': ['mean','max','count','sum']})
```
